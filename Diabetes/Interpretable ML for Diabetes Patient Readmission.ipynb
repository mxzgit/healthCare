```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetic_data.csv')
```

```python
display(df.head(10))
```

Change ? symbole with nan value

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.isnull().sum(axis = 0) / df.shape[0]
```

The columns ['weight','payer_code','medical_specialty'] contain lot of missing values, we will drop this columns at the first step, after we can see if these feature will have an impact on the models we will implement.

```python
df.drop(['weight','payer_code','medical_specialty'],axis=1,inplace=True)
```

Drop the columns related to ids as they will not be used as feature to learn a model

```python
df.drop(['encounter_id','patient_nbr','admission_type_id','discharge_disposition_id','admission_source_id']
       ,axis=1, inplace=True)
```

Remove unkown gender, they are 3 examples of Unknown/Invalid gender

```python
df = df[df.gender != 'Unknown/Invalid']
```

```python
df.dropna(inplace=True)
```

```python
display(df.head(10))
```
